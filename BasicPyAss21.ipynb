# 1. Add the current date to the text file today.txt as a string.

Sure, here's how to add the current date as a string to the "today.txt" file using Python:

```python
import datetime

# get the current date as a string
date_str = datetime.date.today().isoformat()

# open the file in append mode
with open('today.txt', 'a') as file:
    # write the date string to the file
    file.write(date_str + '\n')
```

This will open the "today.txt" file in append mode, write the current date (as a string) followed by a newline character, and then close the file. If the file doesn't exist, it will be created.

# 6. Use multiprocessing to create three separate processes. Make each one wait a random number of seconds between one and five, print the current time, and then exit.

Here's an example implementation using the `multiprocessing` module in Python:

```python
import multiprocessing
import time
import random
from datetime import datetime

def worker():
    wait_time = random.randint(1, 5)
    time.sleep(wait_time)
    print(f"{datetime.now()}: Process {multiprocessing.current_process().name} waited for {wait_time} seconds.")
    
if __name__ == '__main__':
    for i in range(3):
        p = multiprocessing.Process(target=worker, name=f"Process-{i}")
        p.start()
```

In this example, we define a `worker` function that each process will execute. The `worker` function waits for a random number of seconds between 1 and 5, prints the current time and the process name, and then exits.

In the main section of the code, we create three separate processes using a loop and the `Process` constructor from the `multiprocessing` module. We set the `target` argument to `worker`, which is the function that each process will execute. We also set the `name` argument to a unique name for each process.

Finally, we start each process by calling the `start()` method. The `if __name__ == '__main__'` block is used to ensure that the code is only executed when the script is run directly, and not when it is imported as a module.

# 9. When will you be (or when were you) 10,000 days old?

To determine when you will be or were 10,000 days old, you can use the `timedelta` function from the datetime module to add 10,000 days to your date of birth and find the resulting date.

Here's an example code snippet that calculates when you will be or were 10,000 days old:

``` python
import datetime

# Enter your birthdate in YYYY-MM-DD format
birthdate = datetime.date(1995, 10, 15)

# Calculate the date when you will be/were 10,000 days old
ten_thousand_days = datetime.timedelta(days=10000)
ten_thousand_days_date = birthdate + ten_thousand_days

print("You will be/were 10,000 days old on:", ten_thousand_days_date)
```

This will output the date when you will be or were 10,000 days old, depending on whether your birthdate is in the future or the past.
