````python
# 1. Write a Python program to convert kilometers to miles?

# take input for distance in kilometers
kilometers = float(input("Enter distance in kilometers: "))

# convert kilometers to miles using the conversion factor
conv_fac = 0.621371
miles = kilometers * conv_fac

# print the converted value in miles
print(kilometers, "kilometers is equal to", miles, "miles.")
````

````python
# 2. Write a Python program to convert Celsius to Fahrenheit?

# take input temperature in Celsius
celsius = float(input("Enter temperature in Celsius: "))

# convert Celsius to Fahrenheit using the conversion formula
fahrenheit = (celsius * 1.8) + 32

# print the converted temperature in Fahrenheit
print(celsius, "Celsius is equal to", fahrenheit, "Fahrenheit.")
````

````python
# 3. Write a Python program to display calendar?

import calendar

# take input for month and year
year = int(input("Enter the year: "))
month = int(input("Enter the month: "))

# display the calendar for the specified month and year
print(calendar.month(year, month))

````

````python
# 4. Write a Python program to solve quadratic equation?

import math

# take input for coefficients of the quadratic equation
a = float(input("Enter coefficient of x^2: "))
b = float(input("Enter coefficient of x: "))
c = float(input("Enter constant term: "))

# calculate the discriminant
discriminant = (b ** 2) - (4 * a * c)

# check if the roots are real or complex
if discriminant < 0:
    print("The roots are complex.")
else:
    # calculate the roots of the quadratic equation
    root1 = (-b + math.sqrt(discriminant)) / (2 * a)
    root2 = (-b - math.sqrt(discriminant)) / (2 * a)
    
    # print the roots of the quadratic equation
    print("The roots of the quadratic equation are:", root1, "and", root2)
````

# 5. Write a Python program to swap two variables without temp variable?

Here's a Python program to swap two variables without using a temporary variable:
```
# take input for variables a and b
a = int(input("Enter value of a: "))
b = int(input("Enter value of b: "))

# swap the values of a and b
a, b = b, a

# print the swapped values of a and b
print("After swapping, the value of a is:", a)
print("After swapping, the value of b is:", b)
```
In this program, the user is prompted to enter the values of the variables `a` and `b`. The values of `a` and `b` are swapped using a single line of code: `a, b = b, a`. This line of code assigns the value of `b` to `a` and the value of `a` to `b`. Finally, the program prints a message to the console that displays the swapped values of `a` and `b`.
