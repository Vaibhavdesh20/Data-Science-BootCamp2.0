# 1. Make a class called Thing with no contents and print it. Then, create an object called example from this class and also print it. Are the printed values the same or different?



Sure! Here's an example code in Python that defines a class called `Thing` and creates an object of that class called `example`:

```python
class Thing:
    pass

print(Thing())
example = Thing()
print(example)
```

The output of this code will be something like:

```
<__main__.Thing object at 0x7f9d9d1a1be0>
<__main__.Thing object at 0x7f9d9d1a1c50>
```

The printed values are different because they represent two separate instances of the `Thing` class, even though the class itself has no contents. The first printed value is the result of calling `print(Thing())`, which creates a new instance of the `Thing` class and immediately prints its default string representation (which includes the class name and the memory address where the object is stored). The second printed value is the result of calling `print(example)`, which prints the default string representation of the `example` object, which is also an instance of the `Thing` class.

# 3. Make yet another class called, of course, Thing3. This time, assign the value 'xyz' to an instance(object) attribute called letters. Print letters. Do you need to make an object from the class to do this?

Sure! Here's an example code in Python that defines a class called `Thing3` and sets an instance attribute called `letters` with the value `'xyz'`:

```python
class Thing3:
    def __init__(self):
        self.letters = 'xyz'

example = Thing3()
print(example.letters)
```

The output of this code will be `'xyz'`, which is the value of the `letters` instance attribute of the `example` object. We need to create an instance of the `Thing3` class (in this case, the `example` object) in order to access the instance attribute. Each instance of a class can have its own set of instance attributes with different values.

Note that we defined a special method called `__init__` in the `Thing3` class. This method gets called when we create a new instance of the class and allows us to set the initial state of the instance attributes. In this case, we set the `letters` attribute of the instance to `'xyz'`.

# 4. Create an Element class with the instance attributes name, symbol, and number. Create a class object with the values 'Hydrogen,' 'H,' and 1.

Sure! Here's an example code in Python that defines an `Element` class with the instance attributes `name`, `symbol`, and `number`, and creates an object of that class with the values 'Hydrogen,' 'H,' and 1:

```python
class Element:
    def __init__(self, name, symbol, number):
        self.name = name
        self.symbol = symbol
        self.number = number

hydrogen = Element('Hydrogen', 'H', 1)
print(hydrogen.name)
print(hydrogen.symbol)
print(hydrogen.number)
```

The output of this code will be:
```
Hydrogen
H
1
```

In this code, we defined a class called `Element` with an `__init__` method that sets the instance attributes `name`, `symbol`, and `number` based on the arguments passed to the constructor. We then created an instance of the `Element` class called `hydrogen` with the values `'Hydrogen'`, `'H'`, and `1`, respectively, and printed the values of its instance attributes using dot notation.

We then printed the values of its instance attributes using dot notation, which should be the same as before.

# 8. Modify Element to make the attributes name, symbol, and number private. Define a getter property for each to return its value.

Sure! Here's an example code in Python that modifies the `Element` class to make its `name`, `symbol`, and `number` attributes private and defines getter properties for each of them:

```python
class Element:
    def __init__(self, name, symbol, number):
        self.__name = name
        self.__symbol = symbol
        self.__number = number
        
    @property
    def name(self):
        return self.__name
    
    @property
    def symbol(self):
        return self.__symbol
    
    @property
    def number(self):
        return self.__number
```

In this code, we used double underscores (`__`) to make the `name`, `symbol`, and `number` attributes of the `Element` class private. This means that they cannot be accessed or modified from outside the class without using getter and setter methods.

We then defined getter methods for each of the private attributes using the `@property` decorator. These methods have the same names as the private attributes, but without the double underscores. They simply return the value of their respective private attributes.

Note that this implementation only defines getter properties for the private attributes. If you also want to define setter methods to modify them from outside the class, you can use the `@property_name.setter` decorator to define a setter method with the same name as the getter property.

We then created one object from each class and printed what they eat using the `print()` function. The output should be:

```
The bear eats: berries
The rabbit eats: clover
The octothorpe eats: campers
```

# 10. Define these classes: Laser, Claw, and SmartPhone. Each has only one method: does(). This returns 'disintegrate' (Laser), 'crush' (Claw), or 'ring' (SmartPhone). Then, define the class Robot that has one instance (object) of each of these. Define a does() method for the Robot that prints what its component objects do.

Sure! Here's an example code in Python that defines the `Laser`, `Claw`, and `SmartPhone` classes with their respective `does()` methods, and then defines the `Robot` class that has one instance of each of these classes and a `does()` method that prints what its component objects do:

```python
class Laser:
    def does(self):
        return 'disintegrate'

class Claw:
    def does(self):
        return 'crush'

class SmartPhone:
    def does(self):
        return 'ring'

class Robot:
    def __init__(self):
        self.laser = Laser()
        self.claw = Claw()
        self.smartphone = SmartPhone()
        
    def does(self):
        print('The laser:', self.laser.does())
        print('The claw:', self.claw.does())
        print('The smartphone:', self.smartphone.does())

robot = Robot()
robot.does()
```

In this code, we defined three classes `Laser`, `Claw`, and `SmartPhone`, each with their respective `does()` methods that return `'disintegrate'`, `'crush'`, and `'ring'`, respectively. We then defined the `Robot` class that has one instance of each of these classes (`laser`, `claw`, and `smartphone`) as attributes. The `does()` method of the `Robot` class simply calls the `does()` method of each of its component objects and prints what they do.

We then created one object (`robot`) of the `Robot` class and called its `does()` method using the dot notation. The output should be:

```
The laser: disintegrate
The claw: crush
The smartphone: ring
```
