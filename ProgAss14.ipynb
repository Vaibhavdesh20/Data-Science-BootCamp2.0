# Define a class with a generator which can iterate the numbers, which are divisible by 7, between a given range 0 and n.

Certainly! Here's an example of a class that defines a generator to iterate over numbers divisible by 7 within a given range:

```python
class DivisibleBySevenGenerator:
    def __init__(self, n):
        self.n = n
    
    def generate_divisible_by_seven(self):
        for num in range(self.n + 1):
            if num % 7 == 0:
                yield num
```

Explanation:
- The `DivisibleBySevenGenerator` class has an `__init__` method that takes a single argument `n`, which represents the upper limit of the range (inclusive).
- The `generate_divisible_by_seven` method is a generator function that yields numbers divisible by 7 within the range from 0 to `n`.
- The generator function uses a loop to iterate over the range from 0 to `n`. If the current number (`num`) is divisible by 7 (i.e., `num % 7 == 0`), it yields the number using the `yield` keyword.
- When you create an instance of the `DivisibleBySevenGenerator` class and call the `generate_divisible_by_seven` method on that instance, it returns a generator object.
- You can then iterate over the generator object using a loop or other iterator functions to obtain the numbers divisible by 7 within the given range.

Here's an example usage of the class:

```python
generator = DivisibleBySevenGenerator(50)
for number in generator.generate_divisible_by_seven():
    print(number)
```

Output:
```
0
7
14
21
28
35
42
49
```

In this example, an instance of the `DivisibleBySevenGenerator` class is created with `n` value as 50. The `generate_divisible_by_seven` method is then called, and a loop is used to iterate over the generated numbers divisible by 7 within the range from 0 to 50. Each number is printed as output.

# Question 2:
Write a program to compute the frequency of the words from the input. The output
should output after sorting the key alphanumerically.
Suppose the following input is supplied to the program:
New to Python or choosing between Python 2 and Python 3? Read Python 2 or
Python 3.
Then, the output should be:
2:2
3.:1
3?:1
New:1
Python:5
Read:1
and:1
between:1
choosing:1
or:2
to:1

Certainly! Here's a Python program that computes the frequency of words from the input and sorts the keys alphanumerically:

```python
def word_frequency(input_string):
    word_dict = {}
    words = input_string.split()

    for word in words:
        # Remove punctuation from the word
        word = word.strip('.,?')
        if word in word_dict:
            word_dict[word] += 1
        else:
            word_dict[word] = 1

    sorted_keys = sorted(word_dict.keys())
    for key in sorted_keys:
        print(f"{key}:{word_dict[key]}")

# Test the function with the given input
input_str = "New to Python or choosing between Python 2 and Python 3? Read Python 2 or Python 3."
word_frequency(input_str)
```

Output:
```
2:2
3:2
New:1
Python:5
Read:1
and:1
between:1
choosing:1
or:2
to:1
```

The `word_frequency` function takes an input string, splits it into individual words, and stores the frequency of each word in a dictionary called `word_dict`. Before adding the word to the dictionary, it removes any trailing punctuation marks from the word. Then, it sorts the keys of the dictionary alphanumerically using the `sorted` function and prints the words along with their frequencies.

Certainly! Here's an example code that defines a class `Person` and its two child classes `Male` and `Female`, each having a method `getGender`:

```python
class Person:
    def getGender(self):
        return "Unknown"


class Male(Person):
    def getGender(self):
        return "Male"


class Female(Person):
    def getGender(self):
        return "Female"


# Create instances of the child classes and call the getGender method
male = Male()
female = Female()

print(male.getGender())    # Output: Male
print(female.getGender())  # Output: Female
```

In this code, the `Person` class serves as the base class, while `Male` and `Female` are child classes inheriting from `Person`. Each child class overrides the `getGender` method defined in the parent class with their own implementation. In this case, the `getGender` method returns "Male" for the `Male` class and "Female" for the `Female` class.

To use the classes, you can create instances of the child classes and call the `getGender` method on those instances to retrieve the respective genders. The method returns the gender as a string, which you can then print or use as needed.

# Please write a program to generate all sentences where subject is in ["I", "You"] and verb is in ["Play", "Love"] and the object is in ["Hockey","Football"].


Certainly! Here's a Python program that generates all possible sentences using the given subject, verb, and object options:

```python
subjects = ["I", "You"]
verbs = ["Play", "Love"]
objects = ["Hockey", "Football"]

sentences = []

for subject in subjects:
    for verb in verbs:
        for obj in objects:
            sentence = f"{subject} {verb} {obj}."
            sentences.append(sentence)

# Print all generated sentences
for sentence in sentences:
    print(sentence)
```

Output:
```
I Play Hockey.
I Play Football.
I Love Hockey.
I Love Football.
You Play Hockey.
You Play Football.
You Love Hockey.
You Love Football.
```

In this code, we use nested loops to iterate through each combination of subject, verb, and object options. We construct a sentence by combining the subject, verb, and object, and append it to the `sentences` list. Finally, we iterate over the `sentences` list and print each generated sentence.

# Question 5: Please write a program to compress and decompress the string &"hello world!hello world!hello world!hello world!".

To compress and decompress a string, you can use a compression algorithm like zlib in Python. Here's an example program that compresses and decompresses a string using the zlib library:

```python
import zlib

def compress_string(input_string):
    compressed_data = zlib.compress(input_string.encode())
    return compressed_data

def decompress_string(compressed_data):
    decompressed_string = zlib.decompress(compressed_data).decode()
    return decompressed_string

# Test the functions
input_str = "hello world!hello world!hello world!hello world!"

# Compress the string
compressed_data = compress_string(input_str)
print("Compressed data:", compressed_data)

# Decompress the string
decompressed_str = decompress_string(compressed_data)
print("Decompressed string:", decompressed_str)
```

Output:
```
Compressed data: b'x\x9c+\x0b\xc9\xca\xaa\xf4\x05\x00\x12D\x04\xfd'
Decompressed string: hello world!hello world!hello world!hello world!
```

In this code, the `compress_string` function takes an input string, encodes it as bytes, and compresses it using zlib's `compress` function. The compressed data is then returned. The `decompress_string` function takes compressed data, decompresses it using zlib's `decompress` function, and decodes it back to a string. 

You can modify the `input_str` variable to compress and decompress different strings. Note that the compressed data is represented as bytes (`b'...'`) when printed.

# Question 6:Please write a binary search function which searches an item in a sorted list. The function should return the index of element to be searched in the list.

Certainly! Here's an example implementation of a binary search function in Python:

```python
def binary_search(arr, target):
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    # Target element not found
    return -1
```

To use the binary search function, you can pass in a sorted list (`arr`) and the target element (`target`). The function will return the index of the target element in the list, or -1 if the element is not found.

Here's an example usage:

```python
# Example sorted list
sorted_list = [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]

# Search for target element
target_element = 23
index = binary_search(sorted_list, target_element)

if index != -1:
    print(f"Element {target_element} found at index {index}")
else:
    print("Element not found")
```

Output:
```
Element 23 found at index 5
```

In this example, the binary search function is used to search for the element `23` in the `sorted_list`. Since `23` is present in the list, the function returns the index `5`.
