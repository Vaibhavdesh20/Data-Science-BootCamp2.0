```python
# 1. Write a Python Program to Find LCM?

# define a function to find the LCM of two numbers
def lcm(x, y):
    # choose the greater number
    if x > y:
        greater = x
    else:
        greater = y

    # loop through multiples of the greater number until a multiple of both is found
    while True:
        if greater % x == 0 and greater % y == 0:
            lcm = greater
            break
        greater += 1

    return lcm

# take input from user
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

# call the lcm function and print the result
print("The LCM of", num1, "and", num2, "is", lcm(num1, num2))
```

```python
# 2. Write a Python Program to Find HCF?

# define a function to find the HCF of two numbers using the Euclidean algorithm
def hcf(x, y):
    # find the smaller number
    if x < y:
        smaller = x
    else:
        smaller = y

    # loop through possible factors and find the highest common factor
    for i in range(1, smaller+1):
        if x % i == 0 and y % i == 0:
            hcf = i

    return hcf

# take input from user
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

# call the hcf function and print the result
print("The HCF of", num1, "and", num2, "is", hcf(num1, num2))
```

```python
# 3. Write a Python Program to Convert Decimal to Binary, Octal and Hexadecimal?

# take input from user
dec = int(input("Enter a decimal number: "))

# convert to binary, octal, and hexadecimal using built-in functions
bin = bin(dec)
oct = oct(dec)
hex = hex(dec)

# print the results
print("The decimal number", dec, "in binary is", bin)
print("The decimal number", dec, "in octal is", oct)
print("The decimal number", dec, "in hexadecimal is", hex)
```

```python
# 4. Write a Python Program To Find ASCII value of a character?

# take input from user
ch = input("Enter a character: ")

# find the ASCII value using the ord() function
ascii_val = ord(ch)

# print the result
print("The ASCII value of", ch, "is", ascii_val)
```

```python
# 5. Write a Python Program to Make a Simple Calculator with 4 basic mathematical operations?

# define a function for each operation
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    return x / y

# take input from user
print("Select operation:")
print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")

choice = input("Enter choice (1/2/3/4): ")

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

# perform the selected operation
if choice == '1':
    print(num1, "+", num2, "=", add(num1, num2))

elif choice == '2':
    print(num1, "-", num2, "=", subtract(num1, num2))

elif choice == '3':
    print(num1, "*", num2, "=", multiply(num1, num2))

elif choice == '4':
    print(num1, "/", num2, "=", divide(num1, num2))

else:
    print("Invalid input")


```
