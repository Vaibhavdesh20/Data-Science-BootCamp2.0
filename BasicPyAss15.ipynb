The `genPrimes` function starts by creating a list called `primes` that starts with the first prime number, which is 2. It yields this first prime number using the `yield` keyword. 

The function then starts checking odd numbers (starting with 3) for primality. It does this using a `while` loop that runs indefinitely (since we want to keep generating primes as long as the user keeps calling `next()`).

Inside the loop, we first assume that the current number (`num`) is prime, and then we check if it's divisible by any of the previous prime numbers in the `primes` list. If it is, then we know `num` is not prime and we can move on to the next number to check.

If `num` is not divisible by any of the previous primes, then we add it to the `primes` list and yield it using the `yield` keyword.

Finally, we increment `num` by 2 (since we only need to check odd numbers after 2) and continue the loop to check the next number.

To use this generator, you can create an instance of it and then call its `next()` method to generate the next prime number:

```python
prime_generator = genPrimes()

print(next(prime_generator))  # prints 2
print(next(prime_generator))  # prints 3
print(next(prime_generator))  # prints 5
print(next(prime_generator))  # prints 7
print(next(prime_generator))  # prints 11
# and so on...
``` 

Note that since this generator function runs indefinitely, you should have some way to stop generating primes when you've generated enough. You could either manually stop the loop by breaking out of it at some point, or you could use some other mechanism to limit the number of primes generated (such as using `itertools.islice` to slice the generator after a certain number of primes).
