# Q1. What is a probability distribution, exactly? If the values are meant to be random, how can you predict them at all?

A probability distribution is a mathematical function that describes the likelihood of different outcomes in a random event. It maps each possible outcome of a random event to a probability, which is a number between 0 and 1 that represents the chance of that outcome occurring.

While the individual outcomes of a random event cannot be predicted with certainty, the probability distribution allows us to make predictions about the overall behavior of the event. For example, if we know the probability distribution of a coin flip, we can predict that over a large number of flips, approximately half of them will be heads and half will be tails. 

Probability distributions can take different forms depending on the characteristics of the random event being modeled. Some common probability distributions include the normal distribution, the uniform distribution, and the Poisson distribution.

# Q2. Is there a distinction between true random numbers and pseudo-random numbers, if there is one? Why are the latter considered “good enough”?

Yes, there is a distinction between true random numbers and pseudo-random numbers. True random numbers are generated from a physical process that is inherently random, such as atmospheric noise or radioactive decay, while pseudo-random numbers are generated by algorithms that use a deterministic process to produce a sequence of numbers that appear to be random.

While true random numbers are ideal for many applications, they can be difficult to generate and may not be practical for all situations. Pseudo-random numbers, on the other hand, can be generated quickly and easily using mathematical algorithms, and the quality of the sequence can be controlled by the choice of algorithm and seed value. While the sequence of numbers generated by a pseudo-random number generator is not truly random, it can be "good enough" for many purposes, such as simulations or cryptography.

# Q3. What are the two main factors that influence the behaviour of a "normal" probability distribution?

The two main factors that influence the behavior of a normal probability distribution are the mean (μ) and the standard deviation (σ). The mean determines the center of the distribution, while the standard deviation determines the spread or variability of the distribution. A larger standard deviation means a wider distribution, while a smaller standard deviation means a narrower distribution. The shape of the distribution is symmetric, with the highest point at the mean and gradually decreasing on either side.

# Q4. Provide a real-life example of a normal distribution.

A real-life example of a normal distribution is the distribution of human height. In a given population, most people will have a height close to the mean value, with progressively fewer people having heights farther from the mean. The distribution of heights follows a bell curve, with the majority of people having heights in the middle of the distribution and fewer people having heights at the extremes.

# Q5. In the short term, how can you expect a probability distribution to behave? What do you think will happen as the number of trials grows?

In the short term, the behaviour of a probability distribution can be quite variable and unpredictable due to random chance. However, as the number of trials increases, the behaviour of the distribution tends to become more stable and predictable. This is because the effects of random chance tend to average out over large numbers of trials, and the underlying probability distribution becomes more apparent. This is known as the law of large numbers, and it is a fundamental concept in probability theory. Essentially, the more trials you run, the more closely the actual results will approximate the expected outcomes according to the underlying probability distribution.

# Q8. What is the relationship between exponentiation and logarithms?

Exponentiation and logarithms are inverse operations of each other. If we have a number x and raise it to the power of y, we get x^y. If we have x^y = z, then the logarithm of z to the base x is y. Symbolically, we write it as:

log_x(z) = y

Here, x is the base of the logarithm, z is the result of exponentiation, and y is the exponent. 

For example, 2^3 = 8. The logarithm of 8 to the base 2 is 3, because 2 raised to the power of 3 gives 8:

log_2(8) = 3

Similarly, if we take the logarithm of 100 to the base 10, we get 2, because 10 raised to the power of 2 gives 100:

log_10(100) = 2
