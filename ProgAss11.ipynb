```python
# 1. Write a Python program to find words which are greater than given length k?

def find_words_greater_than_length(words, k):
    result = []
    for word in words:
        if len(word) > k:
            result.append(word)
    return result

# Example usage
word_list = ["apple", "banana", "kiwi", "orange", "pear"]
length_threshold = 5
words_greater_than_length = find_words_greater_than_length(word_list, length_threshold)
print("Words greater than length", length_threshold, ":", words_greater_than_length)
```

```python
# 2. Write a Python program for removing i-th character from a string?

def remove_character(string, i):
    if i < 0 or i >= len(string):
        return string

    return string[:i] + string[i+1:]

# Example usage
word = "Hello"
index_to_remove = 2
result = remove_character(word, index_to_remove)
print("Modified string:", result)
```

```python
# 3. Write a Python program to split and join a string?

def split_and_join(string):
    # Split the string into a list of words
    words = string.split()

    # Join the words back into a single string using a space as the delimiter
    joined_string = " ".join(words)

    return joined_string

# Example usage
sentence = "This is a sample sentence"
split_string = split_and_join(sentence)
print("Split string:", split_string)

# Joining a list of strings
words = ["Hello", "World", "!"]
joined_words = " ".join(words)
print("Joined words:", joined_words)
```

```python
# 4. Write a Python to check if a given string is binary string or not?

def is_binary_string(string):
    # Check if all characters in the string are '0' or '1'
    for char in string:
        if char != '0' and char != '1':
            return False
    return True

# Example usage
string1 = "101010"
string2 = "10102"
string3 = "110011"
print("String 1 is binary:", is_binary_string(string1))
print("String 2 is binary:", is_binary_string(string2))
print("String 3 is binary:", is_binary_string(string3))
```

```python
# 5. Write a Python program to find uncommon words from two Strings?

def find_uncommon_words(string1, string2):
    words1 = string1.split()
    words2 = string2.split()

    # Create a set of words from both strings
    set1 = set(words1)
    set2 = set(words2)

    # Find the uncommon words by taking the symmetric difference of the sets
    uncommon_words = set1.symmetric_difference(set2)

    return uncommon_words

# Example usage
string1 = "apple banana mango"
string2 = "banana mango orange"
uncommon_words = find_uncommon_words(string1, string2)
print("Uncommon words:", uncommon_words)
```

```python
# 6. Write a Python to find all duplicate characters in string?

def find_duplicate_characters(string):
    # Create an empty set to store the duplicate characters
    duplicates = set()

    # Create an empty set to store the characters we have seen so far
    seen = set()

    # Iterate over each character in the string
    for char in string:
        # Check if the character is already in the 'seen' set
        # If it is, add it to the 'duplicates' set
        # If it is not, add it to the 'seen' set
        if char in seen:
            duplicates.add(char)
        else:
            seen.add(char)

    return duplicates

# Example usage
input_string = "Hello World"
duplicate_chars = find_duplicate_characters(input_string)
print("Duplicate characters:", duplicate_chars)
```

```python
# 7. Write a Python Program to check if a string contains any special character?

import string

def contains_special_character(string):
    # Define a set of special characters
    special_chars = set(string.punctuation)

    # Check if any special character is present in the string
    for char in string:
        if char in special_chars:
            return True

    return False

# Example usage
input_string = "Hello@World!"
contains_special = contains_special_character(input_string)
print("String contains special character:", contains_special)
```
