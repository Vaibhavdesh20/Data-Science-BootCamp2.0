```python
# 1. Write a Python Program to find sum of array?

def array_sum(arr):
    sum = 0
    for num in arr:
        sum += num
    return sum

# Prompt the user to enter the array elements
elements = input("Enter the array elements separated by spaces: ").split()

# Convert the elements to integers
arr = [int(num) for num in elements]

# Calculate the sum of the array
result = array_sum(arr)

# Display the result
print(f"The sum of the array is: {result}")
```

```python
# 2. Write a Python Program to find largest element in an array?

def find_largest(arr):
    if not arr:
        return None

    largest = arr[0]
    for num in arr:
        if num > largest:
            largest = num
    return largest

# Prompt the user to enter the array elements
elements = input("Enter the array elements separated by spaces: ").split()

# Convert the elements to integers
arr = [int(num) for num in elements]

# Find the largest element in the array
result = find_largest(arr)

# Display the result
if result is None:
    print("The array is empty.")
else:
    print(f"The largest element in the array is: {result}")
```

```python
# 3. Write a Python Program for array rotation?

def rotate_array(arr, k):
    n = len(arr)
    k = k % n  # Adjust k if it's greater than n

    rotated_arr = arr[k:] + arr[:k]
    return rotated_arr

# Prompt the user to enter the array elements
elements = input("Enter the array elements separated by spaces: ").split()

# Convert the elements to integers
arr = [int(num) for num in elements]

# Prompt the user to enter the rotation count
k = int(input("Enter the rotation count: "))

# Perform array rotation
result = rotate_array(arr, k)

# Display the result
print("Array after rotation:", result)
```

```python
# 4. Write a Python Program to Split the array and add the first part to the end?

def split_and_add(arr, k):
    n = len(arr)
    k = k % n  # Adjust k if it's greater than n

    split_arr = arr[k:] + arr[:k]
    return split_arr

# Prompt the user to enter the array elements
elements = input("Enter the array elements separated by spaces: ").split()

# Convert the elements to integers
arr = [int(num) for num in elements]

# Prompt the user to enter the split count
k = int(input("Enter the split count: "))

# Perform splitting and adding
result = split_and_add(arr, k)

# Display the result
print("Array after splitting and adding:", result)
```

```python
# 5. Write a Python Program to check if given array is Monotonic?

def is_monotonic(arr):
    increasing = decreasing = True

    for i in range(1, len(arr)):
        if arr[i] < arr[i-1]:
            increasing = False
        if arr[i] > arr[i-1]:
            decreasing = False

    return increasing or decreasing

# Prompt the user to enter the array elements
elements = input("Enter the array elements separated by spaces: ").split()

# Convert the elements to integers
arr = [int(num) for num in elements]

# Check if the array is monotonic
result = is_monotonic(arr)

# Display the result
if result:
    print("The array is monotonic.")
else:
    print("The array is not monotonic.")
```
