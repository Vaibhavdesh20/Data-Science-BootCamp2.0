# 3. Create a CSV file called books.csv by using these lines:
title,author,year
The Weirdstone of Brisingamen,Alan Garner,1960
Perdido Street Station,China Miéville,2000
Thud!,Terry Pratchett,2005
The Spellman Files,Lisa Lutz,2007
Small Gods,Terry Pratchett,1992

Sure! Here's an example code in Python that creates a CSV file named `books.csv` with the given content:

```python
import csv

header = ['title', 'author', 'year']
rows = [
    ['The Weirdstone of Brisingamen', 'Alan Garner', 1960],
    ['Perdido Street Station', 'China Miéville', 2000],
    ['Thud!', 'Terry Pratchett', 2005],
    ['The Spellman Files', 'Lisa Lutz', 2007],
    ['Small Gods', 'Terry Pratchett', 1992]
]

with open('books.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(header)
    writer.writerows(rows)
```

In this code, we first import the `csv` module, which provides functionality to read and write CSV files.

We then define a list `header` containing the column names, and a list of lists `rows` containing the data.

We open the file `books.csv` in write mode using the `open()` function and the `'w'` mode, and create a `csv.writer` object with the `csv.writer()` function. We also pass `newline=''` to the `open()` function to ensure that line endings are handled correctly.

Inside the `with` block, we call the `writerow()` method of the `csv.writer` object to write the header row, and then call the `writerows()` method to write all the data rows.

This will create a new file named `books.csv` in the current directory (where the Python script is located) with the given content.

# 4. Use the sqlite3 module to create a SQLite database called books.db, and a table called books with these fields: title (text), author (text), and year (integer).

Sure! Here's an example code in Python that creates a SQLite database named `books.db` with a table called `books` with the given fields:

```python
import sqlite3

# Create a connection to the database
conn = sqlite3.connect('books.db')

# Create a cursor object to execute SQL commands
cursor = conn.cursor()

# Create the books table
cursor.execute('''
    CREATE TABLE books (
        title TEXT,
        author TEXT,
        year INTEGER
    )
''')

# Commit the changes to the database and close the connection
conn.commit()
conn.close()
```

In this code, we first import the `sqlite3` module, which provides functionality to work with SQLite databases.

We then create a connection to the database using the `sqlite3.connect()` function, passing the name of the database file as a parameter. If the file doesn't exist, it will be created.

We create a cursor object using the `cursor()` method of the connection object. The cursor object allows us to execute SQL commands on the database.

We use the `execute()` method of the cursor object to execute a CREATE TABLE command that creates a table named `books` with the three fields `title`, `author`, and `year`, using the specified data types.

We then commit the changes to the database using the `commit()` method of the connection object, and close the connection using the `close()` method.

This will create a new SQLite database file named `books.db` in the current directory (where the Python script is located) with a table named `books` with the specified fields.

# 5. Read books.csv and insert its data into the book table.

To read the data from the `books.csv` file and insert it into the `books` table in the `books.db` database using Python and the `sqlite3` module, you can use the following code:

```python
import csv
import sqlite3

# Open the CSV file
with open('books.csv') as file:
    # Create a CSV reader object
    reader = csv.reader(file)
    
    # Skip the header row
    next(reader)
    
    # Create a connection to the database
    conn = sqlite3.connect('books.db')
    
    # Create a cursor object to execute SQL commands
    cursor = conn.cursor()
    
    # Insert the data into the books table
    for row in reader:
        cursor.execute('INSERT INTO books VALUES (?, ?, ?)', row)
    
    # Commit the changes to the database and close the connection
    conn.commit()
    conn.close()
```

In this code, we first import the `csv` and `sqlite3` modules.

We then use the `open()` function to open the `books.csv` file and create a CSV reader object using the `csv.reader()` function. We skip the header row using the `next()` function.

We create a connection to the `books.db` database using the `sqlite3.connect()` function and create a cursor object using the `cursor()` method of the connection object.

We loop through the rows in the CSV file using a for loop and insert each row into the `books` table using an SQL INSERT command with placeholders. We use the `execute()` method of the cursor object to execute the SQL command, passing the row data as a parameter.

We commit the changes to the database using the `commit()` method of the connection object and close the connection using the `close()` method.

# 6. Select and print the title column from the book table in alphabetical order.

To select and print the `title` column from the `book` table in alphabetical order, you can use the following Python code with the `sqlite3` module:

```python
import sqlite3

# Connect to the database and create a cursor
conn = sqlite3.connect('books.db')
cursor = conn.cursor()

# Select the title column from the book table in alphabetical order
cursor.execute('SELECT title FROM books ORDER BY title')

# Fetch all the results and print them
results = cursor.fetchall()
for row in results:
    print(row[0])

# Close the connection
conn.close()
```

In this code, we first import the `sqlite3` module.

We create a connection to the `books.db` database using the `sqlite3.connect()` function and create a cursor object using the `cursor()` method of the connection object.

We then execute an SQL SELECT command to select the `title` column from the `books` table, ordering the results by the `title` column in alphabetical order using the `ORDER BY` clause.

We fetch all the results using the `fetchall()` method of the cursor object and loop through the results, printing each value in the `title` column.

Finally, we close the connection using the `close()` method.
