```python
# 1. Write a Python Program to Find the Factorial of a Number?

def factorial(num):
    if num == 0:
        return 1
    else:
        return num * factorial(num - 1)

# take input from user
num = int(input("Enter a number: "))

# check if the number is negative, zero or positive
if num < 0:
    print("Factorial cannot be calculated for negative numbers.")
elif num == 0:
    print("Factorial of 0 is 1.")
else:
    print("Factorial of", num, "is", factorial(num))
```

```python
# 2. Write a Python Program to Display the multiplication Table?

# take input from user
num = int(input("Enter a number: "))

# display multiplication table
for i in range(1, 11):
    print(num, "x", i, "=", num*i)
```

```python
# 3. Write a Python Program to Print the Fibonacci sequence?

# take input from user
n_terms = int(input("Enter the number of terms: "))

# define the first two terms
first_term, second_term = 0, 1

# check if the number of terms is valid
if n_terms <= 0:
    print("Please enter a positive integer.")
elif n_terms == 1:
    print("Fibonacci sequence upto", n_terms, "term:")
    print(first_term)
else:
    print("Fibonacci sequence upto", n_terms, "terms:")
    for i in range(n_terms):
        print(first_term, end=' ')
        nth_term = first_term + second_term
        # update the values of the first two terms
        first_term = second_term
        second_term = nth_term
```

```python
# 4. Write a Python Program to Check Armstrong Number?

# take input from user
num = int(input("Enter a number: "))

# find the number of digits in the number
num_of_digits = len(str(num))

# initialize the sum variable
sum = 0

# find the sum of the cubes of each digit
temp_num = num
while temp_num > 0:
    digit = temp_num % 10
    sum += digit ** num_of_digits
    temp_num //= 10

# check if the number is an Armstrong number
if num == sum:
    print(num, "is an Armstrong number.")
else:
    print(num, "is not an Armstrong number.")
```

```python
# 5. Write a Python Program to Find Armstrong Number in an Interval?

# take input from user
start_num = int(input("Enter the starting number: "))
end_num = int(input("Enter the ending number: "))

# loop through the interval
for num in range(start_num, end_num + 1):

    # find the number of digits in the number
    num_of_digits = len(str(num))

    # initialize the sum variable
    sum = 0

    # find the sum of the cubes of each digit
    temp_num = num
    while temp_num > 0:
        digit = temp_num % 10
        sum += digit ** num_of_digits
        temp_num //= 10

    # check if the number is an Armstrong number
    if num == sum:
        print(num, "is an Armstrong number.")
```

```python
# 6. Write a Python Program to Find the Sum of Natural Numbers?

# take input from user
n = int(input("Enter a positive integer: "))

# initialize the sum variable
sum = 0

# loop through the natural numbers and add them to the sum
for i in range(1, n+1):
    sum += i

# print the sum of the first n natural numbers
print("The sum of the first", n, "natural numbers is:", sum)
```
