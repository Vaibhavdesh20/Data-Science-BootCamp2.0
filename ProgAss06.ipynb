```python
# 1. Write a Python Program to Display Fibonacci Sequence Using Recursion?

def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)

def display_fibonacci_sequence(count):
    if count <= 0:
        print("Please enter a positive integer.")
    else:
        print("Fibonacci sequence:")
        for i in range(count):
            print(fibonacci(i))

# Prompt the user to enter the count of Fibonacci numbers to display
count = int(input("Enter the count of Fibonacci numbers to display: "))
display_fibonacci_sequence(count)
```

```python
# 2. Write a Python Program to Find Factorial of Number Using Recursion?

def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

# Prompt the user to enter a number
num = int(input("Enter a non-negative integer: "))

if num < 0:
    print("Factorial is not defined for negative numbers.")
else:
    result = factorial(num)
    print(f"The factorial of {num} is {result}.")
```

```python
# 3. Write a Python Program to calculate your Body Mass Index?

def calculate_bmi(weight, height):
    bmi = weight / (height ** 2)
    return bmi

# Prompt the user to enter weight in kilograms
weight = float(input("Enter your weight in kilograms: "))

# Prompt the user to enter height in meters
height = float(input("Enter your height in meters: "))

# Calculate the BMI
bmi = calculate_bmi(weight, height)

# Display the result
print(f"Your Body Mass Index (BMI) is: {bmi:.2f}")
```

```python
# 4. Write a Python Program to calculate the natural logarithm of any number?

import math

# Prompt the user to enter a number
number = float(input("Enter a positive number: "))

# Calculate the natural logarithm
logarithm = math.log(number)

# Display the result
print(f"The natural logarithm of {number} is: {logarithm:.4f}")
```

```python
# 5. Write a Python Program for cube sum of first n natural numbers?

def cube_sum(n):
    sum = 0
    for i in range(1, n+1):
        sum += i ** 3
    return sum

# Prompt the user to enter a positive integer
n = int(input("Enter a positive integer: "))

# Calculate the cube sum
result = cube_sum(n)

# Display the result
print(f"The cube sum of the first {n} natural numbers is: {result}")
```
