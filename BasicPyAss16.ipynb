# 7. Lowercase the last element of the surprise list, reverse it, and then capitalize it.

Sure, I can help you lowercase the last element of `surprise`, reverse it, and then capitalize it.

First, we can use indexing to access the last element in `surprise`, convert it to lowercase using the `lower()` method, and assign the result back to the last element in `surprise`:

```python
surprise[-1] = surprise[-1].lower()
```

This modifies the last element in `surprise` to `"harpo"`.

Next, we can reverse the last element using slicing and the `[::-1]` notation:

```python
surprise[-1] = surprise[-1][::-1]
```

This modifies the last element in `surprise` to `"oprah"`.

Finally, we can capitalize the last element using the `capitalize()` method:

```python
surprise[-1] = surprise[-1].capitalize()
```

This modifies the last element in `surprise` to `"Oprah"`.

To put it all together, we can chain these operations using nested indexing:

```python
surprise[-1] = surprise[-1].lower()[::-1].capitalize()
```

The resulting `surprise` list would be:

```
["Groucho", "Chico", "Oprah"]
```

As you can see, the last element has been converted to lowercase, reversed, and then capitalized.

# 8. Make an English-to-French dictionary called e2f and print it. Here are your starter words: dog is chien, cat is chat, and walrus is morse.

Sure, I can help you create an English-to-French dictionary called `e2f` with the given words and their French translations:

This creates a dictionary called `e2f` with the three word pairs as key-value pairs. We use curly braces to define the dictionary, separate the key-value pairs with colons, and separate the pairs with commas.

The `print()` function is used to display the entire dictionary.

The resulting `e2f` dictionary would be:

```
{'dog': 'chien', 'cat': 'chat', 'walrus': 'morse'}
```

# 10. Make a French-to-English dictionary called f2e from e2f. Use the items method.

Sure, I can help you create a French-to-English dictionary called `f2e` from the existing `e2f` dictionary using the `items()` method. Here's the code:

```python
e2f = {
    "dog": "chien",
    "cat": "chat",
    "walrus": "morse"
}

# Create f2e using the items() method
f2e = {value: key for key, value in e2f.items()}

print(f2e)
```

This uses a dictionary comprehension to create a new dictionary called `f2e` with the values of `e2f` as keys and the keys of `e2f` as values. We use the `items()` method to loop through the key-value pairs in `e2f`.

The resulting `f2e` dictionary would be:

```
{'chien': 'dog', 'chat': 'cat', 'morse': 'walrus'}
```

As you can see, the `f2e` dictionary contains the same word pairs as `e2f`, but with the keys and values reversed.

# 12. Make and print a set of English words from the keys in e2f.

Sure, to make a set of English words from the keys in the `e2f` dictionary, you can use the `set()` function on the dictionary's keys. Here's the code to do it:

```python
e2f = {
    "dog": "chien",
    "cat": "chat",
    "walrus": "morse"
}

# Create a set of English words from the keys in e2f
english_words = set(e2f.keys())

print(english_words)  # prints {'dog', 'cat', 'walrus'}
```

This code creates a set called `english_words` by passing the keys of the `e2f` dictionary to the `set()` function. The resulting set contains the English words from the keys in `e2f`.

Finally, the `print()` function is used to display the set of English words.

# 14. Print the top-level keys of life.

Sure, you can print the top-level keys of the `life` dictionary by calling the `keys()` method on it. Here's the code to do it:

```python
life = {
    "animals": {
        "cats": ["Henri", "Grumpy", "Lucy"],
        "octopi": {},
        "emus": {}
    },
    "plants": {},
    "other": {}
}

# Print the top-level keys of the `life` dictionary
print(life.keys())  # prints dict_keys(['animals', 'plants', 'other'])
```

# 15. Print the keys for life['animals'].

Sure, you can print the keys for the `'animals'` sub-dictionary in the `life` dictionary by using the `'animals'` key to access it, and then calling the `keys()` method on it. Here's the code to do it:
