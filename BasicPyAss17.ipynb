# 1. Assign the value 7 to the variable guess_me. Then, write the conditional tests (if, else, and elif) to print the string 'too low' if guess_me is less than 7, 'too high' if greater than 7, and 'just right' if equal to 7.



Here's the code that will assign the value of 7 to the variable guess_me and print 'too low', 'too high', or 'just right' based on the conditional tests:

```python
guess_me = 7

if guess_me < 7:
    print('too low')
elif guess_me > 7:
    print('too high')
else:
    print('just right')
```

In this code, the `if` statement checks if `guess_me` is less than 7. If it is, the program will execute the code inside the `if` block, which is to print the string 'too low'. 

If `guess_me` is not less than 7, the `elif` statement will check if it is greater than 7. If it is, the program will execute the code inside the `elif` block, which is to print the string 'too high'.

If `guess_me` is neither less than nor greater than 7, then it must be equal to 7. In this case, the `else` block will be executed, which will print the string 'just right'.

# 2. Assign the value 7 to the variable guess_me and the value 1 to the variable start. Write a while loop that compares start with guess_me. Print too low if start is less than guess me. If start equals guess_me, print 'found it!' and exit the loop. If start is greater than guess_me, print 'oops' and exit the loop. Increment start at the end of the loop.


Here's the code that will assign the values of 7 to `guess_me` and 1 to `start`, and then use a `while` loop to compare them:

```python
guess_me = 7
start = 1

while True:
    if start < guess_me:
        print('too low')
    elif start == guess_me:
        print('found it!')
        break
    else:
        print('oops')
        break
    start += 1
```

In this code, the `while` loop will run indefinitely (`while True:`) until it is explicitly broken out of using the `break` statement. 

Inside the loop, an `if` statement checks if `start` is less than `guess_me`. If it is, the program will print the string 'too low'. 

If `start` is equal to `guess_me`, the `elif` statement will be executed, which will print 'found it!' and then break out of the loop using the `break` statement.

If `start` is neither less than nor equal to `guess_me`, the `else` block will be executed, which will print the string 'oops' and then break out of the loop using the `break` statement.

Finally, after each iteration of the loop, `start` is incremented by 1 using the `start += 1` statement.

# 3. Print the following values of the list [3, 2, 1, 0] using a for loop.

Here's the code that will print the values of the list [3, 2, 1, 0] using a `for` loop:

```python
my_list = [3, 2, 1, 0]

for num in my_list:
    print(num)
```

In this code, the `for` loop will iterate over each element in the `my_list` list. The variable `num` is assigned to each element in turn, and then printed using the `print()` function.

The output of this code will be:

```
3
2
1
0
``` 

Each number in the list will be printed on a separate line.

# 4. Use a list comprehension to make a list of the even numbers in range(10)

Here's the code that will use a list comprehension to make a list of the even numbers in the range of 10:

```python
even_nums = [num for num in range(10) if num % 2 == 0]
print(even_nums)
```

In this code, the list comprehension creates a new list called `even_nums` that contains only the even numbers in the range of 10. 

The `for` loop part of the comprehension iterates over each number in the range of 10, while the `if` statement filters out any number that is not even (`num % 2 == 0` checks if the number is divisible by 2).

The output of this code will be:

```
[0, 2, 4, 6, 8]
```

This list contains only the even numbers in the range of 10.

# 5. Use a dictionary comprehension to create the dictionary squares. Use range(10) to return the keys, and use the square of each key as its value.

Here's the code that will use a dictionary comprehension to create a dictionary called `squares` with keys from the range of 10 and their respective values as their squares:

```python
squares = {num: num**2 for num in range(10)}
print(squares)
```

In this code, the dictionary comprehension creates a new dictionary called `squares` by iterating over each number in the range of 10 using the `for` loop. 

The key-value pair for each number is created by specifying `num` as the key and `num**2` as the value. `num**2` calculates the square of the number.

The output of this code will be:

```
{0: 0, 1: 1, 2: 4, 3: 9, 4: 16, 5: 25, 6: 36, 7: 49, 8: 64, 9: 81}
```

This dictionary contains the squares of the numbers in the range of 10 as its values, with each number as its respective key.

# 6. Construct the set odd from the odd numbers in the range using a set comprehension (10).

Here's the code that will use a set comprehension to construct a set called `odd` with only the odd numbers from the range of 10:

This set contains only the odd numbers in the range of 10.

# 7. Use a generator comprehension to return the string 'Got' and a number for the numbers in range(10). Iterate through this by using a for loop.

Here's the code that will use a generator comprehension to return the string 'Got' and a number for the numbers in the range of 10, and then iterate through it using a `for` loop:

```python
gen = ('Got ' + str(num) for num in range(10))

for item in gen:
    print(item)
```

The output of this code will be:

```
Got 0
Got 1
Got 2
Got 3
Got 4
Got 5
Got 6
Got 7
Got 8
Got 9
```

Each item in the generator object is printed on a separate line.

# 9. Define a generator function called get_odds that returns the odd numbers from range(10). Use a for loop to find and print the third value returned.

Here's the code that defines a generator function called `get_odds` that returns the odd numbers from the range of 10, and uses a `for` loop to find and print the third value returned:

```python
def get_odds():
    for num in range(1, 10, 2):
        yield num

count = 1
for odd in get_odds():
    if count == 3:
        print("The third odd number is:", odd)
        break
    count += 1
```

In this code, the `def` keyword is used to define a generator function called `get_odds`. This function uses a `for` loop to iterate over each odd number in the range of 10, and `yield`s each odd number as it is generated.

The `for` loop iterates over the generator object returned by `get_odds()`. Inside the loop, the `if` statement checks if the loop has iterated three times by checking if `count` is equal to 3. If it is, the third odd number is printed using the `print()` function and the loop is exited using the `break` statement. The `count` variable is incremented by 1 for each iteration of the loop.

The output of this code will be:

```
The third odd number is: 5
```

This is the third odd number generated by the `get_odds()` generator function.

# 10. Define an exception called OopsException. Raise this exception to see what happens. Then write the code to catch this exception and print 'Caught an oops'.

Sure, here's the code to define and raise the `OopsException` exception:

```python
class OopsException(Exception):
    pass

raise OopsException
```

When you run this code, it will raise the `OopsException` exception and print a traceback of the error.

To catch the exception and print 'Caught an oops', you can use a try-except block like this:

```python
try:
    raise OopsException
except OopsException:
    print('Caught an oops')
```

This code first tries to raise the `OopsException` exception, but then catches it with the `except` block and executes the print statement. The output of this code will be:

```
Caught an oops
```

# 11. Use zip() to make a dictionary called movies that pairs these lists: titles = ['Creature of Habit','Crewel Fate'] and plots = [A nun turns into a monster,;A haunted yarn shop].


Sure, here's the code to use `zip()` to create a dictionary `movies` that pairs the `titles` and `plots` lists:

```python
titles = ['Creature of Habit', 'Crewel Fate']
plots = ['A nun turns into a monster', 'A haunted yarn shop']

movies = dict(zip(titles, plots))
print(movies)
```

In this code, `zip()` combines the two lists `titles` and `plots` into a single iterable of tuples, where each tuple contains a title and a plot. The `dict()` function then converts this iterable into a dictionary where the titles are the keys and the plots are the values.

The output of this code will be:

```
{'Creature of Habit': 'A nun turns into a monster', 'Crewel Fate': 'A haunted yarn shop'}
```
