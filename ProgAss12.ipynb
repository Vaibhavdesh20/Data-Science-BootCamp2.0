```python
# 1. Write a Python program to Extract Unique values dictionary values?

def extract_unique_values(dictionary):
    # Create an empty list to store unique values
    unique_values = []

    # Iterate over each value in the dictionary
    for value in dictionary.values():
        # If the value is not already in the unique_values list, add it
        if value not in unique_values:
            unique_values.append(value)

    return unique_values

# Example usage
my_dict = {
    "key1": 1,
    "key2": 2,
    "key3": 3,
    "key4": 2,
    "key5": 3
}
unique_values = extract_unique_values(my_dict)
print("Unique values:", unique_values)
```

```python
# 2. Write a Python program to find the sum of all items in a dictionary?

def sum_dictionary_items(dictionary):
    # Get the values from the dictionary and calculate their sum
    total_sum = sum(dictionary.values())

    return total_sum

# Example usage
my_dict = {
    "key1": 10,
    "key2": 20,
    "key3": 30,
    "key4": 40
}
dictionary_sum = sum_dictionary_items(my_dict)
print("Sum of dictionary items:", dictionary_sum)
```

```python
# 3. Write a Python program to Merging two Dictionaries?

def merge_dictionaries(dict1, dict2):
    merged_dict = dict1.copy()  # Create a copy of dict1

    merged_dict.update(dict2)  # Update the copy with the key-value pairs from dict2

    return merged_dict

# Example usage
dict1 = {"key1": "value1", "key2": "value2"}
dict2 = {"key3": "value3", "key4": "value4"}
merged_dict = merge_dictionaries(dict1, dict2)
print("Merged dictionary:", merged_dict)
```

```python
# 4. Write a Python program to convert key-values list to flat dictionary?

def convert_to_flat_dictionary(key_value_list):
    flat_dict = {}

    for pair in key_value_list:
        key, value = pair
        flat_dict[key] = value

    return flat_dict

# Example usage
key_value_list = [("key1", "value1"), ("key2", "value2"), ("key3", "value3")]
flat_dict = convert_to_flat_dictionary(key_value_list)
print("Flat dictionary:", flat_dict)
```

```python
# 5. Write a Python program to insertion at the beginning in OrderedDict?

from collections import OrderedDict

def insert_at_beginning(ordered_dict, key, value):
    # Create a new OrderedDict with the new key-value pair at the beginning
    new_ordered_dict = OrderedDict([(key, value)])

    # Update the new OrderedDict with the remaining key-value pairs from the original OrderedDict
    new_ordered_dict.update(ordered_dict)

    return new_ordered_dict

# Example usage
my_dict = OrderedDict([('key1', 'value1'), ('key2', 'value2')])
key = 'key3'
value = 'value3'
updated_dict = insert_at_beginning(my_dict, key, value)
print("Updated OrderedDict:", updated_dict)
```

```python
# 6. Write a Python program to check order of character in string using OrderedDict()?

from collections import OrderedDict

def check_order_of_characters(string, pattern):
    # Create an OrderedDict to store the occurrences of characters in the string
    char_dict = OrderedDict()

    # Iterate over each character in the string
    for char in string:
        # Add the character to the OrderedDict if it's not already present
        if char not in char_dict:
            char_dict[char] = 1

    # Iterate over each character in the pattern
    for char in pattern:
        # If the character is not present in the OrderedDict or its count is 0, return False
        if char not in char_dict or char_dict[char] == 0:
            return False
        # Decrease the count of the character in the OrderedDict
        char_dict[char] -= 1

    # If all characters in the pattern have been found in the correct order, return True
    return True

# Example usage
string = "Hello World"
pattern = "HrWd"
order_check = check_order_of_characters(string, pattern)
print("Order of characters in the string:", order_check)
```

```python
# 7. Write a Python program to sort Python Dictionaries by Key or Value?

def sort_dict_by_key(dictionary):
    sorted_dict = dict(sorted(dictionary.items(), key=lambda item: item[0]))
    return sorted_dict

def sort_dict_by_value(dictionary):
    sorted_dict = dict(sorted(dictionary.items(), key=lambda item: item[1]))
    return sorted_dict

# Example usage
my_dict = {"key2": 5, "key1": 10, "key3": 3}
sorted_dict_by_key = sort_dict_by_key(my_dict)
sorted_dict_by_value = sort_dict_by_value(my_dict)

print("Sorted dictionary by key:", sorted_dict_by_key)
print("Sorted dictionary by value:", sorted_dict_by_value)
```
