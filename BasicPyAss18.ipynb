# 3. Using the interpreter, explicitly import and call the hours() function from zoo.

Certainly! Here's an example of how to explicitly import and call the `hours()` function from the `zoo` module in the Python interactive interpreter:

```python
>>> from zoo import hours
>>> hours()
Open 9-5 daily
```

In this example, we use the `from` keyword to explicitly import the `hours()` function from the `zoo` module. Then, we call the `hours()` function directly without having to use the `zoo.` prefix, since we have imported it into the current namespace.

# 5. Create a plain dictionary with the key-value pairs 'a': 1, 'b': 2, and &'c': 3, and print it out.

Sure! Here's an example of how to create a plain dictionary with the key-value pairs `'a': 1`, `'b': 2`, and `'c': 3`, and print it out in the Python interactive interpreter:

```python
>>> my_dict = {'a': 1, 'b': 2, 'c': 3}
>>> print(my_dict)
{'a': 1, 'b': 2, 'c': 3}
```

In this example, we create a plain dictionary named `my_dict` using curly braces `{}` and assigning key-value pairs separated by colons `:`. We then print out the dictionary using the `print()` function.

# 6.Make an OrderedDict called fancy from the same pairs listed in 5 and print it. Did it print in the same order as plain?

Sure! Here's an example of how to create an `OrderedDict` named `fancy` with the same key-value pairs as the plain dictionary in the previous question, and print it out in the Python interactive interpreter:

```python
>>> from collections import OrderedDict
>>> fancy = OrderedDict([('a', 1), ('b', 2), ('c', 3)])
>>> print(fancy)
OrderedDict([('a', 1), ('b', 2), ('c', 3)])
```

In this example, we first import the `OrderedDict` class from the `collections` module. Then, we create an `OrderedDict` named `fancy` using the `OrderedDict()` constructor, passing in a list of key-value pairs as tuples. We then print out the `fancy` dictionary using the `print()` function.

As an `OrderedDict` maintains the order in which the key-value pairs were added, the output should be the same order as we added them to the dictionary, which is `{'a': 1, 'b': 2, 'c': 3}` in this case.

# 7. Make a default dictionary called dict_of_lists and pass it the argument list. Make the list dict_of_lists['a'] and append the value 'something for a' to it in one assignment. Print dict_of_lists['a'].

Sure! Here's an example of how to create a `defaultdict` called `dict_of_lists`, make a new list for key `'a'`, append a value to it in one assignment, and print the resulting list:

```python
>>> from collections import defaultdict
>>> dict_of_lists = defaultdict(list)
>>> dict_of_lists['a'].append('something for a')
>>> print(dict_of_lists['a'])
['something for a']
```

In this example, we first import the `defaultdict` class from the `collections` module. We then create a `defaultdict` named `dict_of_lists` with the default value of an empty list using the `defaultdict()` constructor.

Next, we access the `'a'` key of `dict_of_lists` using square brackets `[]`, which creates a new empty list because it does not exist yet. We then append the value `'something for a'` to the newly created list in the same assignment using the `.append()` method.

Finally, we print the resulting list for the `'a'` key using `print(dict_of_lists['a'])`. The output should be `['something for a']`.
