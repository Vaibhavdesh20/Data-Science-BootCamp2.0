Q5. Why do you no longer use string-based exceptions?

String-based exceptions were deprecated in Python 2.x and removed in Python 3.x. Instead of using string-based exceptions, Python now uses classes to define exceptions. There are several reasons why string-based exceptions were removed:

1. Lack of clarity: 

String-based exceptions provided limited information about the error that occurred. They were often generic and did not provide much context to help diagnose the issue.

2. No hierarchy: 

With string-based exceptions, there was no clear hierarchy to the exceptions. This made it difficult to handle exceptions in a structured way and forced developers to write custom code to handle each exception individually.

3. No type checking:

With string-based exceptions, there was no way to perform type checking on exceptions. This made it difficult to catch specific types of exceptions and required developers to manually parse the exception message to determine the type of exception.

4. Error-prone: 

Using string-based exceptions could be error-prone. Developers could easily mistype an exception name or include typos in the exception message, which could make it difficult to diagnose the issue.

By using classes to define exceptions, Python provides a more structured and robust way of handling errors. Classes allow for a clear hierarchy of exceptions, type checking, and provide additional context information to help diagnose issues. Additionally, classes can be extended and customized, allowing developers to create their own exceptions to handle specific use cases.
